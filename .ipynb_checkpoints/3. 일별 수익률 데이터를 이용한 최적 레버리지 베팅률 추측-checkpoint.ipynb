```python
# 포트폴리오 수익률을 이용하여 기댓값이 어느정도 나올지 추측해본다.
# 참고 : https://datascienceschool.net/view-notebook/dd6a7633d69f401bb00409b9ae8806e8/

## 필요 패키지 호출

import numpy as np
import pandas as pd
import matplotlib as mpl
```

```python
## 수익률 파일 호출
## 총자산 및 남은 현금 Columns는 필요성이 느껴지지 않아 Drop하였음.

daily_profit = pd.read_csv('Test_CSV/913695/trade_history_daily_913695.csv')
daily_profit = daily_profit.drop(["총자산", "남은 현금"], axis=1)

print(daily_profit)
```

```python
##  일별수익률의 평균 및 표준편차 추출

port_avg = round(daily_profit["일일수익률"].mean(), 3)
port_std = round(daily_profit["일일수익률"].std(), 3)

print(port_avg, port_std)
```

```python
## 일별 누적평균 및 누적 표준편차확인

expanding_port_avg = daily_profit["일일수익률"].expanding(2).mean()
expanding_port_std = daily_profit["일일수익률"].expanding(2).std()

print(expanding_port_avg)
print(expanding_port_std)
```

```python
## 그래프로 시각화하여 누적평균 확인

import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

plt.style.use('fivethirtyeight')

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

expanding_port_avg.plot(figsize=(6,4))
```

```python
## 그래프로 시각화하여 누적 표준편차 확인

expanding_port_std.plot(figsize=(6,4))
```

```python
## 수익률 정규분포 확인
import seaborn as sns

sns.distplot(daily_profit["일일수익률"])
plt.title("포트폴리오 일별 수익률 분포")
plt.xlabel("일간수익률")
plt.show() 
```

```python
## 승률 통계내기
win = daily_profit["일일수익률"][daily_profit["일일수익률"] > 0].count()
tot = len(daily_profit.index)
print(round(win/tot, 3))
```

```python
## 전날 수익률과 오늘 수익률의 관계는 어떻게 될까?
## 상수 K : 수익이 난 날 전일의 총 일 횟수
win_list = daily_profit["일일수익률"][daily_profit["일일수익률"] > 0].index.tolist()
past_avg_list = []
k = 2

## 반복문을 이용하여 당일 수익률, D-1일 수익률 예측
for i in win_list[k : len(win_list)]:
    print(i, daily_profit["일일수익률"][i], round(daily_profit["일일수익률"][i-k:i-1].mean(), 3))
    
### 2020.02.17
### 수익난 날의 과거 데이터를 이용하여 수익률을 극대화 할 수 있는 방안을 연구해본다.
```

```python
## DataFrame으로 만든 후 피어슨 상관계수를 이용해 상관관계를 확인해본다.
import math

temp_array = []

for i in win_list[k:len(win_list)]:
    today_profit = daily_profit["일일수익률"][i]
    past_profit = round(daily_profit["일일수익률"][i-(k+1):i-1].mean(), 3) 
    
    ## 데이터가 없을 경우 0으로 처리
    if math.isnan(past_profit) is True:
        past_profit = 0
    
    temp_tuple = (i, today_profit, past_profit)
    temp_array.append(temp_tuple)
       
## DataFrame로 저장 후 피어슨 상관계수 분석
profit_flow_df = pd.DataFrame(temp_array, columns = ["index", "today_profit", "past_profit"]) 
corr = profit_flow_df[["today_profit", "past_profit"]].corr()
print(corr)
```

```python
## 그래프로 시각화하여 확인

fig, ax = plt.subplots()
ax.plot(profit_flow_df["index"], profit_flow_df[["today_profit", "past_profit"]])

ax.set(xlabel='index', ylabel='profit',title='profit_analyze')
ax.grid()

plt.show()
```
